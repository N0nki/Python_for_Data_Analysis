## P17　Python標準機能を使用したタイムゾーン情報の集計
タイムゾーンのうち最も頻度の高いものを調べる

```python
# タイムゾーンがないレコードもあるのでキー'tz'を含むレコードのみ処理する
time_zones = [rec['tz'] for rec in records if 'tz' in rec]

# 最初の５件を表示
for i in range(5):
    print(time_zones[i])
```

```python
from collections import defaultdict

# 集計を行う関数を定義
# キーがタイムゾーン、値がタイムゾーンの出現回数の辞書を返す
def get_counts(sequence):
    counts = defaultdict(int) # 辞書のデフォルト値を0に設定
    for x in sequence:
        counts[x] += 1
    return counts
```

```python
len(time_zones)
```

```python
counts = get_counts(time_zones)
```

```python
counts['America/New_York']
```

上位１０件のタイムゾーンを調べる

```python
# 方法1 組み込み関数のみ
def top_counts(count_dict, n=10):
    # key,valueを取り出す 詳しくはdictionary view objects
    value_key_pairs = [(count, tz) for tz,count in count_dict.items()]
    # sort()は破壊的変更を行う
    # シーケンス型を要素とするリストに対しては
    # インデックス番号0の要素に'<'を適用して昇順ソート
    value_key_pairs.sort()
    return value_key_pairs[-n:]

# 方法2 collections.Counter
from collections import Counter

top_counts(counts)
c = Counter(time_zones)
c.most_common(10)
```

ユーザエージェント文字列の先頭トークンを切り出してまとめて表示する

```python
results = Series([x.split()[0] for x in frame.a.dropna()])
results[:5]
```

```python
results.value_counts()[:10]
```

レコード文字列に'Windows'が含まれるかを判定する

```python
# np.where(condition,x,y)
# conditionが真のときx、偽のときy
operating_system = np.where(frame['a'].str.contains('Windows'), 'Windows', 'Not WIndows')
for s in operating_system[:5]:
    print(s)
```

frameにgroupbyを適応して、frameのレコードをタイムゾーンと稼働OSの組み合わせごとにグループ分けする

```python
by_tz_os = frame.groupby(['tz', operating_system])
by_tz_os
```

```python
agg_counts = by_tz_os.size().unstack().fillna(0)
agg_counts[:10]
```

タイムゾーン情報全体から上位のものを見つける  
まずagg_countsをタイムゾーンの出現頻度別に集計する

```python
indexer = agg_counts.sum(1).argsort()
indexer[:10]
```

上記の結果は例えば２レコード目に着目すると、ソートした結果が２番目であるレコードはagg_countsのインデックス番号２０番目の要素であるということを表している。  
indexerの一列目は、そのタイムゾーンがagg_countsでは何行目のレコードだったのかを示している。

```python
count_subset = agg_counts.take(indexer)[-10:]
count_subset
```

```python
count_subset.plot(kind='barh', stacked=True, grid=True)
plt.show()
```

Windowsユーザの相対的な割合を読み取れるようにするために、各行の合計が１となるように正規化する。

```python
normed_subset = count_subset.div(count_subset.sum(1), axis=0)
normed_subset.plot(kind='barh', stacked=True, grid=True)
```

MovieLens 1Mは100万件の映画評価データで、4000本の映画に対する6000人の評価をまとめたものである。データは3つの表から構成されており、それぞれ評価、ユーザ情報、映画情報となっている。  
データについての詳細な情報はpydata-book/ch02/movielens/READMEを参照。

pandasの`read_table`をつかってデータをDataFrameオブジェクトとして読み込む。  
本書のサンプルコードのままではセパレータを正しく扱えずエラーがでる。これを回避するには`read_table`にキーワード引数を`endine='python'`を追加すればよい。  

参考
* [Python file path failing in pycharm regex confusion](http://stackoverflow.com/questions/27301477/python-file-path-failing-in-pycharm-regex-confusion)
* [Read csv with “^” + character delimiter with the C engine](http://stackoverflow.com/questions/25204642/read-csv-with-character-delimiter-with-the-c-engine)

```python
import pandas as pd

unames = ['user_id', 'gender', 'age', 'occupation', 'zip-code']
users = pd.read_table('pydata-book/ch02/movielens/users.dat', sep='::', header=None, names=unames, engine='python')

rnames = ['user_id', 'movie_id', 'rating', 'timestanp']
ratings = pd.read_table('pydata-book/ch02/movielens/ratings.dat', sep='::', header=None, names=rnames, engine='python')

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('pydata-book/ch02/movielens/movies.dat', sep='::', header=None, names=mnames, engine='python')
```

3つの表を1つにまとめる。まずratingとusersを結合し、次にその結果とmoviesを結合する。この操作で使用するpandasのmergeは2つの表に共通列が存在するとき、その列をキーとして結合することができる。

```python
data = pd.merge(pd.merge(ratings, users), movies)
data
```

# P34　データ分析の例：アメリカの赤ちゃんに名付けられた名前リスト（1880-2010）

アメリカ合衆国社会保障局の提供する赤ちゃんの名前データを分析する。  
* ある特定の名前の出現頻度が、年代順にどのように移り変わるのかを視覚的に表示する
* ある名前の相対頻度を求める
* 年ごとに、最も人気のあった名前や、前年と比べて最も増加した名前、もしくは減少した名前を求める
* 名付けの傾向を分析してみる：どんな母音が多いのか、どんな子音が多いのか。また多様性はどうか。スペルが変化しているか。最初の文字で多いものは何か。末尾の文字で多いものは何か。
* 傾向の外部要因を分析してみる：聖書に出てくる名前はどうか。嫁偉人の名前はどうか。別の観点で、人口動態の変化はわかるか。

`pandas.read_csv`を使ってデータを読み込む

```python
names1880 = pd.read_csv('pydata-book/ch02/names/yob1880.txt', names=['name', 'sex', 'birth'])
names1880
```

birth列をsex列で場合分けして足し合わせて、各年度ごとの性別出生数の概数を得る。

```python
names1880.groupby('sex')['birth'].sum()
```

年度ごとに別れているデータをすべて統合し1つのデータフレームとする。  
新たな列として年度を追加する。  
pandas.concatを利用する。このメソッドは列の追加ではなく行の追加を行う。

```python
years = range(1880, 2011)

pieces  = []
columns = ['name', 'sex', 'births']

for year in years:
    path = "pydata-book/ch02/names/yob%d.txt" % year
    frame = pd.read_csv(path, names=columns)
    
    frame['year'] = year
    pieces.append(frame)

# 各年度のデータフレームを統合
# オリジナルの行番号は不要なのでignore_index=Trueとする
names = pd.concat(pieces, ignore_index=True)
names
```

```python
total_births = pd.pivot_table(names, values='births', index=['year'], columns=['sex'], aggfunc=sum)
total_births
```

```python
total_births.plot(title='Total births by sex and year')
```

データフレームに新たな列propを追加する。propは全出生数に対するその名前の割合を表す。  
例えばpropが0.02であるということは、赤ちゃん100人がいたとしたら、その名前の赤ちゃんは2人いることを示している。  
このpropを年度別・性別ごとに計算していく。

```python
names.groupby(['year', 'sex']).size()
```

```python
def add_prop(group):
    births = group.births
    group['prop'] = births / births.sum()
    return group
names = names.groupby(['year', 'sex']).apply(add_prop)
names
```

各グループを足し合わせて1になるかを確認する。

```python
np.allclose(names.groupby(['year', 'sex']).prop.sum(), 1)
```

年代・性別ごとの、上位1000件の名前がどのようなものか見ていく。

```python
def get_top1000(group):
    return group.sort_values(by='births', ascending=False)[:1000]

grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)
top1000
```
